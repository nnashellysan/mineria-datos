### Data Transform

```python
df_dropped = df.dropna(subset=['MPI Urban'])
df_dropped.head()
```

```python
df_dropped.shape
```

```python
df_copy=df.copy()
```

##### Data Visualization

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```python
df_sample['Intensity of Deprivation Rural']=df_sample['Intensity of Deprivation Rural'].astype('int64')
```

```python
df_sample.plot(kind='scatter',x='Intensity of Deprivation Rural',y='Intensity of Deprivation Urban')
```

```python
plt.hist(df_sample['Intensity of Deprivation Rural'])
plt.xlabel('Intensity of Deprivation Rural')
plt.ylabel('Frequency')
plt.title('Frequency of Intensity of Deprivation Rural')
```

```python
plt.boxplot(df_sample['Frequency of Intensity of Deprivation Rural'], showmeans=True, meanline=True)
plt.ylabel('Frequency of Intensity of Deprivation Rural')
plt.title('Box Plot of Frequency of Intensity of Deprivation Rural')
```
